<h1><b>Without Encoding Technique

```python
df1.head()
```

```python
# applying encode in sex column without encoding technique
df1.sex.unique()
```

```python
# replace male = 1 & female = 2
df1.sex = df.sex.replace(['male', 'female'],[1,2])
df1.head()
```

```python
# applying encode in smoker column without encoding technique
df1.smoker.unique()
```

<h1><b>Label Encoding Technique</h1></b>

```python
from sklearn.preprocessing import LabelEncoder  # import LabelEncoder module
le = LabelEncoder()  # create a object of LabelEncoder is le
```

```python
df2.head()
```

```python
# label encoding on sex column
df2.sex = le.fit_transform(df2.sex)
df2.head()
```

```python
# label encoding on smoker column
df2.smoker = le.fit_transform(df2.smoker)
df2.head()
```

```python
# label encoding on region column
df2.region = le.fit_transform(df2.region)
df2.head()
```

```python
#Label Encoder using for Loop
df3.head()
```

```python
import numpy as np  # import numpy library
import warnings as wr 
wr.filterwarnings('ignore')

for column in df3.columns:
    if df3[column].dtype == np.number:
        continue
    df3[column] = le.fit_transform(df3[column])
```

```python
df3.head()  # after encoding using loop
```

<h1><b>Another Way to Encode

```python
df4.head()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
for column in df4.columns:
    if is_numeric_dtype(df4[column]):
        continue
    else:
        df4[column] = le.fit_transform(df4[column])
```

```python
df4.head()
```

<h1><b>Ordinal Encoding Technique

Ordinal Encoding in sex column

```python
df8.sex.unique()
```

```python
gender = ['female', 'male']

from sklearn.preprocessing import OrdinalEncoder
oe_gen = OrdinalEncoder(categories=[gender])
```

```python
oe_gen
```

```python
encoded_gen = oe_gen.fit_transform(df9[['sex']])
```

```python
en_gen = pd.DataFrame(encoded_gen,columns=['gender'])
```

```python
en_gen.head()
```

```python
df8.head()
```

```python
cols = ['sex']

for col in cols:
    unique_gen = df9[col].unique()
    df9[col] = OrdinalEncoder(categories=[unique_gen]).fit_transform(df9[[col]])
```
